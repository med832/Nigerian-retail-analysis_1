```python
monthly_revenue = (
    df
    .groupby('month')
    ['total_revenue']
    .sum()
)
```

```python
(monthly_revenue / 1e9).round(2)
```

```python
(monthly_revenue / 1e9).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title(
'Monthly Revenue (Billion NGN)'
)

plt.xlabel(
'Month'
)

plt.xticks(rotation=0)

plt.ylabel(
'Revenue (Billion)'
)

plt.show()
```

```python
city_revenue = (
    df
    .groupby(
        'city'
    )
    ['total_revenue']
    .mean()
    .sort_values(
        ascending=False
    )
)
```

```python
city_revenue
```

```python
city_revenue.plot(
    kind='bar',
    figsize=(12,5)
)

plt.title(
'Average Revenue by City'
)

plt.xlabel(
'City'
)

plt.ylabel(
'Average Revenue'
)

plt.xticks(rotation=45)

plt.show()
```
